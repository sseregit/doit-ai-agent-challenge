```python
from langchain_community.tools import TavilySearchResults
from dotenv import load_dotenv

load_dotenv()

tavily_search = TavilySearchResults(
    max_results=5
)
```

```python
res = tavily_search.invoke({"query": "25-26시즌 웨스트브룩 전망"})
for r in res:
  print(r)
```

```python
from IPython.display import JSON
JSON(res).data
```

```python
from tavily import TavilyClient

client = TavilyClient()

content = client.search(
    "25-26시즌 웨스트브룩 전망",
    search_depth="advanced",
    include_raw_content=True,
)

JSON(content).data
```

```python
from langchain.adapters.openai import convert_openai_messages
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-5-2025-08-07", temperature=1)

query = "25-26시즌 웨스트브룩 전망"

content = client.search(
    query,
    include_raw_content=True,
    search_depth="advanced"
)["results"]
```

```python
prompt = [{
  "role": "system",
  "content": (
    "당신은 신문 기사를 쓰는 기자 AI입니다. \n"
    "당신은 주어진 정보를 바탕으로 객관적이고 체계적으로 작성된 기사를 써야 합니다. \n"
  )
}, {
  "role": "user",
  "content": (
    f'정보: """{content}"""\n\n'
    f'위 정보를 사용하여, 다음 질문에 대해 자세한 보고서를 한국어로 작성하세요.: "{query}"\n'
    '-신문 기사 형식으로 사용하되, MLA 표준을 준수하는 markdown 문법을 사용해주세요.'
    '-활용한 자료는 출처를 명시하세요.'
  )
}]
```

```python
report = llm.invoke(prompt).content
print(report)
```
